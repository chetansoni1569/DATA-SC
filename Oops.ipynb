```python
class Complex:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Complex(self.x + other.x, self.y + other.y)

    def display(self):
        print(self.x, " +i", self.y)


a = Complex(2, 3)
b = Complex(4, 5)

c = a + b

print(c.x, c.y)
c.display()
```
