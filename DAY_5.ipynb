```python
import numpy as np

#create a 5*5 matrix of random integer (1-90)
```

```python
mat = np.random.randint(1, 91, size=(5, 5))
print("matrix: ",mat)


#find the mean value row and col wise
row_wisw_mean=np.mean(mat,axis=1)
col_wisw_mean=np.mean(mat,axis=0)

print("rowwise mean: ", row_wisw_mean)
print("colwise mean: ", col_wisw_mean)

#replace  all even number with -1;

modified_mat=np.where(mat%2==0,-1,mat)
print("modified matrix: ",modified_mat)

#sort the array row and column wise
sorted_row_wise=np.sort(mat,axis=1)
sorted_col_wise=np.sort(mat,axis=0)
print("sorted row wise: ",sorted_row_wise)
print("sorted col wise: ",sorted_col_wise)

#find row and colum wise sum seperately
row_wise_sum=np.sum(mat,axis=1)
col_wise_sum=np.sum(mat,axis=0)

print("row wise sum: ",row_wise_sum)
print("col wise sum: ",col_wise_sum)

#count the number of negative in each column
count_col =np.sum(np.where(mat<0,1,0), axis=0)
print(count_col)

print("number of negative in each column: ",count_col)

```

```python
import numpy as np
#create a 5*5 array (stock level)
# create another 5*5 array (demand)

arr1= np.random.randint(1, 100, size=(5, 5))
arr2= np.random.randint(1, 100, size=(5, 5))

print("arr1: ",arr1)
print("arr2: ",arr2)

#compute:
#remaning stock= stock-demand
remaning_stock= arr1-arr2
print("remaning_stock: ",remaning_stock)


#replace :
#nagative values ->0 (out of stock)
remaning_stock[remaning_stock<0]=0
print("remaning_stock: ",remaning_stock)

#find:
#totoal shortage
total_shortage= np.sum(remaning_stock[remaning_stock<0])
print("total_shortage: ",total_shortage)

#products with zero stock
zero_stock= np.where(remaning_stock==0)
print("zero_stock: ",zero_stock)

#sort each row based on remaning stock
sorted_remaning_stock= np.sort(remaning_stock, axis=1)
print("sorted_remaning_stock: ",sorted_remaning_stock)

```
