```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#line chart
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.show()

plt.figure(figsize=(1,1))
plt.title('relation btw number and squars')
plt.tight_layout()
plt.show()
plt.plot(x,y)
```

```python
subjects=['maths','physics','cs','english']
scores=[88,76,95,70]
plt.figure(figsize=(3,3))
plt.bar(subjects,scores,color='green',edgecolor='black')
plt.xlabel('subjects')
plt.ylabel('scores')
plt.title('bar chart')
plt.show()
```

```python
hours= np.random.randint(1,10,50)
marks=hours*8+np.random.randint(1,20,50)
plt.scatter(hours,marks)
```

```python
a=np.random.randint(1,20,50)
b=np.random.randint(10,30,50)
plt.scatter(a,b,s=8)
```

```python
data= np.random.randint(10,80,50)
plt.hist(data,bins=3,edgecolor='black',color='green')
plt.title('distrinut')
plt.xlabel('score')
plt.ylabel('frequency')
plt.show()
```

```python
subjects= ['maths','physics','cs','english']
scores=[30,40,10,20]


plt.pie(scores,labels=subjects,autopct='%.2f%%')
plt.tight_layout()
plt.show()
```

```python
batch1=np.random.randint(10,70,100)
plt.boxplot(batch1,label='batch1')
plt.tight_layout()
plt.show()
```
